```python
f = open('p042_words.txt')
w = f.readline()
w = w.split(",")
words = []
for i in w:
    words.append(i[1:-1])
```

```python
words
```

```python
# compute alphabetic value of words
#print(sum(map(ord,"SKY"))-len("SKY")*64)
values = []
for word in words:
    v = sum(map(ord,word))-len(word)*64
    values.append(v)
```

```python
values
```

```python
# find longest word
longest = 0
for w in words:
    longest = max(longest,len(w))
```

```python
longest
```
