```python
parts = [1]
ks = []
for i in range(1,51):
    ks.append(i)
    ks.append(-i)
```

```python
gks = []
for i in range(100):
    gks.append(int(ks[i]*(3*ks[i]-1)/2))
```

```python
import math
```

```python
parts = [1]
for n in range(1,101):
    tmp = 0
    index = 0
    fac = -1
    while(gks[index] <= n):
       
        if (index %2 == 0):
            fac *= -1
        tmp += ( fac * parts[n-gks[index]])
        print(n,index,fac,tmp)
        index += 1
    parts.append(int(tmp))
    print(parts)
```
